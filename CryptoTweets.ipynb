```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/rodincode/CryptoTweets/main/DOGE-INR.csv",index_col="Date",parse_dates=['Date'])
df = df.drop(columns=["Volume"])
df.head()
```

```python
ax = df.plot(figsize=(15,7))
ax
```

```python
dft1 = pd.read_csv("https://raw.githubusercontent.com/rodincode/CryptoTweets/main/ElonMuskTweets2020.csv")
dft1 = pd.concat([dft1["date"],dft1["tweet"]],axis = 1)
dft1.head()
```

```python
dft2 = pd.read_csv("https://raw.githubusercontent.com/rodincode/CryptoTweets/main/ElonMuskTweets2021.csv")
dft2 = pd.concat([dft2["date"],dft2["tweet"]],axis = 1)
dft2.head()  #2010-06-04 till 2021-03-22
```

```python
dogedf = dft2[dft2['tweet'].str.contains("Dogecoin|dogecoin|DOGECOIN|DogeCoin")]
dogedf
```

```python
doge = dogedf["date"]
dogedates=[]
for i in doge:
  date = i.split()[0]
  print(date)
  dogedates.append(date)
```

```python
cryptweets = df[df.index.isin(dogedates)]
cryptweets
```

```python
cryptweets.plot(style=['+-','o-','.--','s:',".-"],figsize=(15,7))
```

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

plt.style.use("fivethirtyeight")

fig, ax = plt.subplots(figsize=(15,7))

ax.plot(df.index, df["Open"],linewidth = 2, alpha = .7, label='Open')
ax.legend(loc="upper left") 
ax.plot(df.index, df["High"],linewidth = 2, alpha = .7, label='High') 
ax.legend(loc="upper left")       
ax.plot(df.index, df["Low"],linewidth = 2, alpha = .7, label='Low')
ax.legend(loc="upper left")        
ax.plot(df.index, df["Close"],linewidth = 2, alpha = .7, label='Close')
ax.legend(loc="upper left") 
ax.plot(df.index, df["Adj Close"],linewidth = 2, alpha = .7, label='Adj Close')   
ax.legend(loc="upper left")     

ax.plot(cryptweets.index, cryptweets["Open"],"bo", label="Musk's Tweets")
ax.legend(loc="upper left") 
ax.set(xlabel="Date",
       title="DOGECOIN on Elon Musk's Tweet")
# Format the x axis
ax.xaxis.set_major_locator(mdates.WeekdayLocator(interval=10))
ax.xaxis.set_major_formatter(mdates.DateFormatter("%m-%d-%y"))


ax.annotate('Elon Musk Tweeted',xy=(mdates.date2num(cryptweets.index[0]),2.712202),xytext=(-150, 90), 
            textcoords='offset points', arrowprops=dict(arrowstyle='fancy',facecolor="black"))
ax.xaxis.zoom(0.5)
plt.show()
```
