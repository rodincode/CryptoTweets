```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/rodincode/CryptoTweets/main/DOGE-INR.csv",index_col="Date")
df = df.drop(columns=["Volume"])
df.head()
```

```python
ax = df.plot(figsize=(15,7))
ax
```

```python
dft1 = pd.read_csv("https://raw.githubusercontent.com/rodincode/CryptoTweets/main/ElonMuskTweets2020.csv")
dft1 = pd.concat([dft1["date"],dft1["tweet"]],axis = 1)
dft1.head()
```

```python
dft2 = pd.read_csv("https://raw.githubusercontent.com/rodincode/CryptoTweets/main/ElonMuskTweets2021.csv")
dft2 = pd.concat([dft2["date"],dft2["tweet"]],axis = 1)
dft2.head()  #2010-06-04 till 2021-03-22
```

```python
dogedf = dft2[dft2['tweet'].str.contains("Dogecoin|dogecoin|DOGECOIN|DogeCoin")]
dogedf
```

```python
doge = dogedf["date"]
dogedates=[]
for i in doge:
  date = i.split()[0]
  print(date)
  dogedates.append(date)
```

```python
cryptweets = df[df.index.isin(dogedates)]
cryptweets
```

```python
import matplotlib.pyplot as plt
fig, ax1 = plt.subplots(figsize=(15,7))
ax1.plot(df.index, df, 'b-')       
ax1.plot(cryptweets.index, cryptweets,'r*')
plt.show()
```
