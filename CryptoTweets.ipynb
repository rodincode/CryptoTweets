```python
import pandas as pd
dfd = pd.read_csv("https://raw.githubusercontent.com/rodincode/CryptoTweets/main/Dogecoin%20Historical%20Data%20-%20Investing.com%20India.csv",index_col="Date",parse_dates=['Date'])
dfd = dfd.drop(columns=["Vol."])
dfb = pd.read_csv("https://raw.githubusercontent.com/rodincode/CryptoTweets/main/Bitcoin%20Historical%20Data%20-%20Investing.com%20India.csv",index_col = "Date",parse_dates = ["Date"])
dfb = dfb.drop(columns=["Vol.","Change %"])
# if you want to operate on multiple columns, put them in a list like so:
dfb.replace(',','', regex=True, inplace=True)
dfb = dfb.astype(float)
dfd.head()
dfb.head()
```

```python
axd = dfd.plot(figsize=(15,7))
axd
```

# BitCoin Plot

```python
axb = dfb.plot(figsize=(15,7))
axb
```

# Musk's Tweets

```python
dft1 = pd.read_csv("https://raw.githubusercontent.com/rodincode/CryptoTweets/main/ElonMuskTweets2020.csv")
dft1 = pd.concat([dft1["date"],dft1["tweet"]],axis = 1)
dft1.head()
```

```python
dft2 = pd.read_csv("https://raw.githubusercontent.com/rodincode/CryptoTweets/main/ElonMuskTweets2021.csv")
dft2 = pd.concat([dft2["date"],dft2["tweet"]],axis = 1)
dft2.head()  #2010-06-04 till 2021-03-22
```

```python
dogedf = dft2[dft2['tweet'].str.contains("Dogecoin|dogecoin|DOGECOIN|DogeCoin|BitCoin|bitcoin|BITCOIN|Bitcoin")]
dogedf
```

```python
doge = dogedf["date"]
dogedates=[]
for i in doge:
  date = i.split()[0]
  print(date)
  dogedates.append(date)
```

```python
cryptweets1 = dfd[dfd.index.isin(dogedates)]
cryptweets1
```

```python
cryptweets2 = dfb[dfb.index.isin(dogedates)]
cryptweets2
```

```python
cryptweets1.plot(style=['+-','o-','.--','s:',".-"],figsize=(15,7))
```

```python
cryptweets2.plot(style=['+-','o-','.--','s:',".-"],figsize=(15,7))
```

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

plt.style.use("fivethirtyeight")

fig, ax = plt.subplots(figsize=(15,7))

ax.plot(dfd.index, dfd["Open"],linewidth = 2, alpha = .7, label='Open')
ax.legend(loc="upper left") 
ax.plot(dfd.index, dfd["High"],linewidth = 2, alpha = .7, label='High') 
ax.legend(loc="upper left")       
ax.plot(dfd.index, dfd["Low"],linewidth = 2, alpha = .7, label='Low')
ax.legend(loc="upper left")        
# ax.plot(df.index, df["Close"],linewidth = 2, alpha = .7, label='Close')
# ax.legend(loc="upper left") 
# ax.plot(df.index, df["Adj Close"],linewidth = 2, alpha = .7, label='Adj Close')   
# ax.legend(loc="upper left")     

ax.plot(cryptweets.index, cryptweets1["Open"],"bo", label="Musk's Tweets")
ax.legend(loc="upper left") 
ax.set(xlabel="Date",
       title="DOGECOIN on Elon Musk's Tweet on Bitcoin and Dogecoin")
# Format the x axis
ax.xaxis.set_major_locator(mdates.WeekdayLocator(interval=10))
ax.xaxis.set_major_formatter(mdates.DateFormatter("%m-%d-%y"))


ax.annotate('Elon Musk Tweeted',xy=(mdates.date2num(cryptweets.index[0]),2.712202),xytext=(-150, 90), 
            textcoords='offset points', arrowprops=dict(arrowstyle='fancy',facecolor="black"))
ax.xaxis.zoom(0.5)
plt.show()
```
